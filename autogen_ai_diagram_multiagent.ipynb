```python

ragproxyagent = RetrieveUserProxyAgent(
    name="ragproxyagent",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=3,
    retrieve_config={
        "task": "code",
        "docs_path": [
            "https://diagrams.mingrammer.com/docs/getting-started/examples",
            "https://diagrams.mingrammer.com/docs/guides/diagram",
            "https://diagrams.mingrammer.com/docs/guides/node",
            "https://diagrams.mingrammer.com/docs/guides/cluster",
            "https://diagrams.mingrammer.com/docs/guides/edge",
            "https://gist.githubusercontent.com/Praveen-Babu-techie/741eba2927f28e63bdc851bb946536d7/raw/e5bea8df7453e1f4b2dd87f7bedff95c5d905563/azure-diagram.txt"
        ],
        "chunk_token_size": 2000,
        "model": config_list[0]["model"],
        "vector_db": "chroma",
        "overwrite": True,  # Changed to True to ensure a new collection is created
        "get_or_create": True,
    },
    # code_execution_config=False
     is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("TERMINATE"),
     code_execution_config={
        # the executor to run the generated code
        "executor": LocalCommandLineCodeExecutor(work_dir="coding"),
    },
)
```

```python
def handle_design_query(user_input):
  code_problem = f"{user_input} write the script using python diagrams module, with direction as LR in a separate file and execute the script, \n At the end of the script have a print statement the Script executed scuccessfully "
  # print(code_problem)
  chat_result = ragproxyagent.initiate_chat(
      assistant, message=ragproxyagent.message_generator, problem=code_problem, search_string="diagrams"
  )

  verbose_str="*"*20+"AGENT EXECUTION "+"*"*20+"\n"
  img_name=""
  cwd=os.getcwd()+"/coding"
  files=glob.glob(cwd+"/*.png")
  for file in files:
        # img_name=file.split("/")[-1]
        img_name=file
        break



  for i in chat_result.chat_history:


    verbose_str+="-"*100+"\n"
    verbose_str+="ROLE "+i["role"].upper()+"\n"

    verbose_str+="-"*100+"\n"
    verbose_str+="NAME "+i["name"].upper()+"\n"

    verbose_str+="-"*100+"\n"

    verbose_str+="CONTENT "+i["content"]+"\n"
  return verbose_str,img_name
```

```python
import gradio as gr
from PIL import Image

# Function to fetch or generate image and text based on input
def display_content(text_input):
    clear_assistant()
    v_data, image_url = handle_design_query(text_input)
    return [gr.update(visible=True, value=v_data),
            gr.update(visible=True, value=image_url)]

# Create the Gradio interface with conditional visibility
with gr.Blocks() as iface:

    gr.Markdown("# Design Diagram Generator")
    # gr.Markdown("Enter your prompt with corresponding services and flows.")

    with gr.Row():
        # Left column for all text elements
        with gr.Column(scale=1):
            input_text = gr.Textbox(
                label="Enter your prompt with corresponding services and flows.",
                placeholder="Type here...",
                lines=3,
                max_lines=40
            )
            output_text = gr.Textbox(
                label="Agent Verbose",
                lines=5,
                interactive=False,
                visible=False  # Initially hidden
            )
            submit_btn = gr.Button("Generate")

        # Right column for image
        with gr.Column(scale=1):
            output_image = gr.Image(
                label="Generated Design Image",
                type="filepath",
                visible=False  # Initially hidden
            )

    # Connect the components
    submit_btn.click(
        fn=display_content,
        inputs=input_text,
        outputs=[output_text, output_image]
    )

# Launch the app
iface.launch(share=True)
```

```python
# filename: azure_game_architecture.py

from diagrams import Diagram
from diagrams.azure.compute import AppServices, FunctionApps, ContainerInstances
from diagrams.azure.database import BlobStorage, CosmosDb
from diagrams.azure.analytics import Databricks
from diagrams.azure.network import TrafficManagerProfiles, CDNProfiles
from diagrams.azure.integration import APIManagement
from diagrams.azure.mobile import NotificationHubs

with Diagram("Azure Game Architecture", show=False, direction="LR") as diag:
    traffic_manager = TrafficManagerProfiles("Traffic Manager")
    app_service = AppServices("App Service")
    function_app = FunctionApps("Function App")
    container_instance = ContainerInstances("Container Instance")
    api_management = APIManagement("API Management")
    cdn = CDNProfiles("CDN")
    blob_storage = BlobStorage("Blob Storage")
    cosmos_db = CosmosDb("Cosmos DB")
    databricks = Databricks("Databricks")
    notification_hub = NotificationHubs("Notification Hub")

    traffic_manager >> [app_service, function_app, container_instance, api_management]
    cdn >> blob_storage
    traffic_manager >> cdn
    app_service >> cosmos_db
    function_app >> cosmos_db
    container_instance >> cosmos_db
    databricks >> cosmos_db
    databricks >> function_app
    function_app >> notification_hub

print("Script executed successfully")
diag
```
